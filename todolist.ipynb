```python
#Introduction:
# This is a console-based To-Do List application.
# You can add tasks, view pending/all tasks, mark tasks complete, and delete tasks.
# Tasks are stored in a MySQL database, so the data stays even after closing the app.

import pymysql

# Connect Python to MySQL
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="yourpassword",   # replace with your MySQL password
    database="todo_app"         # database we created in MySQL
)

# Create a cursor
cursor = conn.cursor()

print("Connected to MySQL database successfully!")
```

```python
# Function: Add a New Task
def add_task(task_name):
    query = "INSERT INTO tasks (task_name) VALUES (%s)"
    cursor.execute(query, (task_name,))
    conn.commit()
    print(f"Task '{task_name}' added successfully!")
```

```python
# Function: View Pending Tasks
def view_pending_tasks():
    query = "SELECT task_id, task_name, status FROM tasks WHERE status='Pending'"
    cursor.execute(query)
    rows = cursor.fetchall()
    if rows:
        print("\nPending Tasks:")
        for row in rows:
            print(f"{row[0]}. {row[1]} [{row[2]}]")
    else:
        print("\nNo pending tasks!")
```

```python
# Function: View All Tasks
def view_all_tasks():
    query = "SELECT task_id, task_name, status FROM tasks"
    cursor.execute(query)
    rows = cursor.fetchall()
    if rows:
        print("\nAll Tasks:")
        for row in rows:
            print(f"{row[0]}. {row[1]} [{row[2]}]")
    else:
        print("\nNo tasks found!")
```

```python
# Function: Mark Task Complete
def mark_complete(task_id):
    query = "UPDATE tasks SET status='Completed' WHERE task_id=%s"
    cursor.execute(query, (task_id,))
    conn.commit()
    if cursor.rowcount:
        print(f"Task {task_id} marked as completed!")
    else:
        print(f"Task {task_id} not found!")
```

```python
# Function: Delete Task
def delete_task(task_id):
    query = "DELETE FROM tasks WHERE task_id=%s"
    cursor.execute(query, (task_id,))
    conn.commit()
    if cursor.rowcount:
        print(f"Task {task_id} deleted successfully!")
    else:
        print(f"Task {task_id} not found!")
```

```python
# This loop provides a simple menu so the user can interact with the app.
while True:
    print("\n--- To-Do List Menu ---")
    print("1. Add Task")
    print("2. View Pending Tasks")
    print("3. View All Tasks")
    print("4. Mark Task Complete")
    print("5. Delete Task")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        task_name = input("Enter task name: ")
        add_task(task_name)
    elif choice == '2':
        view_pending_tasks()
    elif choice == '3':
        view_all_tasks()
    elif choice == '4':
        task_id = input("Enter task ID to mark complete: ")
        mark_complete(task_id)
    elif choice == '5':
        task_id = input("Enter task ID to delete: ")
        delete_task(task_id)
    elif choice == '6':
        print("Exiting To-Do List App. Goodbye!")
        break
    else:
        print("Invalid choice! Please enter a number from 1-6.")
```
